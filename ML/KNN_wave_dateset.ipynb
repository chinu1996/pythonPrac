```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn 
%matplotlib inline
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
```

```python


#KNN for regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#Instantiate the model
reg = KNeighborsRegressor(n_neighbors=3)

#fit the model 
reg.fit(X_train, y_train)
```

```python
#Make prediction on the test set
print("Test set predictions: \n{}".format(reg.predict(X_test)))
```

```python
#R^2 is called coefficient of determination, measures between 0 and 1
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,4))
#create 100, data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1,1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    #make prediction using 1, 3, 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0),markersize=8)
    ax.plot(X_test, y_test, '^', c=mglearn.cm2(1),markersize=8)
    
    ax.set_title(
        "{}  neighbor(s)\n train_score{:.2f} test_score{:.2f}".format(
            n_neighbors, reg.score(X_train, y_train),
            reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions","Training data/target","Test data/target"], loc="best")
    
```
